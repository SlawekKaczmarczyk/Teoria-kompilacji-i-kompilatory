```python
import lexer_parser as lexer_parser
lexer = lexer_parser.lexer
parser = lexer_parser.parser
from state import State
from lang_types import *

def run_code(code,state=None):
    result = parser.parse(lexer.input(code))
    if state != None:
        result.grant_state(state)
    result.evaluate()
    
    return result.state


# code = """
#     (1+1)*2/3+(1);
#     """
# state = run_code(code)

code = """
    var x <- 1;
    """
state = run_code(code)

code = """
    var x <- "qweasdzxc";
    """
state = run_code(code)

code = """
    var x <- true;
    """
state = run_code(code)

code = """
    var x <- "a";
    var y <- false;
    var z <- null;
    var w <- 345;
    """
state = run_code(code)
assert state.namespaces['root']['x'].value == "a"
assert type(state.namespaces['root']['x']) == String
assert state.namespaces['root']['y'].value == False
assert type(state.namespaces['root']['y']) == Bool
assert state.namespaces['root']['z'].value == None
assert type(state.namespaces['root']['z']) == Null
assert state.namespaces['root']['w'].value == 345
assert type(state.namespaces['root']['w']) == Number

code = """
    if true {
        print 2;
    };
    """
state = run_code(code)

code = """
    var x <- 1;
    if true {
        var x <- 0;
        print "this value should be 0: "+ (x).STR();
    };
    print x;
    """
state = run_code(code)

code = """
    if false {
        print "this shouldn't appear (if false)";
    };
    """
state = run_code(code)

code = """
    if true {
        print "this should appear (if true)";
    };
    """
state = run_code(code)


code = """
        var x <- 5;
        print "this value should be 5: " + x.STR();
    """
state = run_code(code)

code = """
        1.PLUS(2);
        1.MINUS(2);
        1.DIVIDE(2);
        1.TIMES(2);
        ("ASD").PLUS("ZXC");
    """
state = run_code(code)


code = """
        print 1;
    """
state = run_code(code)

code = """
        print 1;
    """
state = run_code(code)

code = """
   print 1 = 1;
   var x <- "asd";
   var y <- "asd";
   print x=y;
   
   """
state = run_code(code)

code = """
   var i <- 5;
   var x <= i;
   
   print "should be 5: " + x.REF().STR();
   
   """
state = run_code(code)

code = """
   var i <- 5;
   var x <= i;
   
   x.SET(0);
   
   print "should be 0: " + i.STR();
   
   
   """
state = run_code(code)

code = """
   var i <- 5;
   var x <= i;
   i <- 2;
   
   
   print "should be 2: " + x.REF().STR();
   
   """
state = run_code(code)

code = """
   var i <- 5;
   
   until true {
   print "odliczanie :" + i.STR();
   i <- i-1;
   };
   
   
   """
state = run_code(code)

code = """
  print 0 = 0;
  print 0;
  print "should be true: " +(0 = 2 -3+1).STR();
  
  print 2 - 3 + 1;
   
   
   """
state = run_code(code)


```
