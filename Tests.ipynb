**Console can be cluttered with print statements, no error means it's fine**

```python
import lexer_parser as lexer_parser
lexer = lexer_parser.lexer
parser = lexer_parser.parser
from state import State
from lang_types import *

def run_code(code,state=None):
    result = parser.parse(lexer.input(code))
    if state != None:
        result.grant_state(state)
    result.evaluate()
    
    return result.state


code = """
    (1+1)*2/3+(1);
    """
state = run_code(code)

code = """
    var x <- 1;
    """
state = run_code(code)

code = """
    var x <- "qweasdzxc";
    """
state = run_code(code)

code = """
    var x <- true;
    """
state = run_code(code)

code = """
    var x <- "a";
    var y <- false;
    var z <- null;
    var w <- 345;
    """
state = run_code(code)
assert state.namespaces['root']['x'].value == "a"
assert type(state.namespaces['root']['x']) == String
assert state.namespaces['root']['y'].value == False
assert type(state.namespaces['root']['y']) == Bool
assert state.namespaces['root']['z'].value == None
assert type(state.namespaces['root']['z']) == Null
assert state.namespaces['root']['w'].value == 345
assert type(state.namespaces['root']['w']) == Number

code = """
    if true {
        print 2;
    };
    """
state = run_code(code)

code = """
    var x <- 1;
    if true {
        var x <- 0;
        print x;
    };
    print x;
    """
state = run_code(code)

code = """
    if true {
        print 2;
    };
    """
state = run_code(code)

code = """
    if false {
        print 2;
    };
    """
state = run_code(code)

code = """
    if true {
        print 2;
    };
    """
state = run_code(code)


code = """
        var x <- 5;
        print x;
    """
state = run_code(code)



```
