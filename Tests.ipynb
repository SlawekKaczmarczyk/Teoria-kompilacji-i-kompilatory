```python
import lexer_parser as lexer_parser
from state import State
from lang_types import *

def run_code(code,state=None):
    
    lexer = lexer_parser.lexer
    parser = lexer_parser.parser
    result = parser.parse(lexer.input(code))
    if state != None:
        result.grant_state(state)
    result.evaluate()
    
    return result.state

def SEP________(c):
    print(f"{c} --------------------")




code = """
    var x <- 1;
    """
state = run_code(code)

SEP________(1)

code = """
    var x <- "qweasdzxc";
    """
state = run_code(code)

SEP________(2)

code = """
    var x <- true;
    """
state = run_code(code)

SEP________(3)

code = """
    var x <- "a";
    var y <- false;
    var z <- null;
    var w <- 345;
    """
state = run_code(code)
assert state.namespaces['root']['x'].value == "a"
assert type(state.namespaces['root']['x']) == String
assert state.namespaces['root']['y'].value == False
assert type(state.namespaces['root']['y']) == Bool
assert state.namespaces['root']['z'].value == None
assert type(state.namespaces['root']['z']) == Null
assert state.namespaces['root']['w'].value == 345
assert type(state.namespaces['root']['w']) == Number

SEP________(4)

code = """
    if true {
        print 2;
    };
    """
state = run_code(code)

SEP________(5)

code = """
    var x <- 1;
    if true {
        var x <- 0;
        print "this value should be 0: "+ (x).STR();
    };
    print x;
    """
state = run_code(code)

SEP________(6)

code = """
    if false {
        print "this shouldn't appear (if false)";
    };
    """
state = run_code(code)

SEP________(7)

code = """
    if true {
        print "this should appear (if true)";
    };
    """
state = run_code(code)

SEP________(8)


code = """
        var x <- 5;
        print "this value should be 5: " + x.STR();
    """
state = run_code(code)

SEP________(9)

code = """
        1.PLUS(2);
        1.MINUS();
        1.DIVIDE(2);
        1.TIMES(2);
        ("ASD").PLUS("ZXC");
    """
state = run_code(code)

SEP________(10)

code = """
        print 1;
    """
state = run_code(code)

SEP________(11)

code = """
   print 1 = 1;
   var x <- "asd";
   var y <- "asd";
   print x=y;
   """
state = run_code(code)

SEP________(12)

code = """
   var i <- 5;
   var x <= i;
   
   print "should be 5: " + x.REF().STR();
   """
state = run_code(code)

SEP________(13)

code = """
   var i <- 5;
   var x <= i;
   
   x.SET(0);
   
   print "should be 0: " + i.STR();
   """
state = run_code(code)

SEP________(14)

code = """
   var i <- 5;
   var x <= i;
   i <- 2;
   
   print "should be 2: " + x.REF().STR();
   """
state = run_code(code)

SEP________(15)

code = """
  print "should be true: " +(0 = 2 -3+1).STR();
   """
state = run_code(code)

SEP________(16)

code = """
  var x <- 0 << 0;
  print "this should be false: " + x.STR();
   
   """
state = run_code(code)

SEP________(17)

code = """
  var x <- 0 << 4;
  print "this should be true: " + x.STR();
   
   """
state = run_code(code)

SEP________(18)

code = """
  var x <- 2 >> 8;
  print "this should be false: " + x.STR();
   
   """
state = run_code(code)

SEP________(19)

code = """
  var x <- 80 >> 8;
  print "this should be true: " + x.STR();
   
   """
state = run_code(code)

SEP________(20)

code = """
  var x <- 2;
  var y <- 1+1;
  
  print x = y;
   
   """
state = run_code(code)

SEP________(21)

code = """
   var i <- 5;
   
   until i = 0 {
   print "odliczanie 5 do 1 :" + i.STR();
   i <- i-1;
   };
   
   """
state = run_code(code)

SEP________(22)

code = """
    fun foo(){
       print "inside function";
       };
    foo();

"""
state = run_code(code)

SEP________(23)

code = """
    var x <- "outside value";
    fun foo(){
       print x;
       };
    foo();

"""
state = run_code(code)

SEP________(24)

code = """
    var x <- 0;
    var ref <= x;
    
    print x;
    print ref;
    fun foo(x){
        print x;
        print x.REF();
        x.SET(3);
       };
    foo(ref);
    
    print x;

"""
state = run_code(code)

SEP________(25)

code = """
    var x <- 0;
    var ref <= x;
    
    print x;
    print ref;
    fun foo(x){
        print x;
        print x.REF();
        x.SET(3);
       };
    foo(ref);
    
    print x;
    

"""
state = run_code(code)

SEP________(25)

code = """
   print 5 % 2;

"""
state = run_code(code)









```
